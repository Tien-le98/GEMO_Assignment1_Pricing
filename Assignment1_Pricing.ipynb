```python
def calculatePrice1():
    base_price_coffee = 2
    drink_type = input("Which drink type would you like? - hot, cold, blended: ")
    while drink_type.lower() != "hot" and drink_type.lower() != "cold" and drink_type.lower() != "blended":
        drink_type = input("Please choose a valid drink type - hot, cold, blended: ")
    if drink_type.lower() == "hot":
        drink_size = input("Which size would you like? - S, M: ")
        while drink_size.upper() != "S" and drink_size.upper() != "M":
            drink_size = input("Please choose a valid drink size - S, M: ")
        if drink_size.upper() == "M":
            base_price_coffee += 0.5
    else:
        drink_size = input("Which size would you like? - S, M, L")
        while drink_size.upper() != "S" and drink_size.upper() != "M" and drink_size.upper() != "L":
            drink_size = input("Please choose a valid drink size - S, M, L: ")
        if drink_type.lower() == "blended":
            base_price_coffee += 1
        if drink_size.upper() == "M":
            base_price_coffee += 0.5
        elif drink_size.upper() == "L":
            base_price_coffee += 1
    whipped_cream = input("Do you want to add whipped cream to your drink? - Y for yes, N for no: ")
    while whipped_cream.upper() != "Y" and whipped_cream.upper() != "N":
        whipped_cream = input("Please choose a valid option for adding whipped cream - Y for yes, N for no: ")
    if whipped_cream.upper() == "Y":
        base_price_coffee += 0.5
    return base_price_coffee
```

```python
def calculatePrice3():
    extra_charge = 0
    chocolate_sauce = input("Do you want to add chocolate sauce? - Y for yes, N for no: ")
    while chocolate_sauce.upper() != "Y" and chocolate_sauce.upper() != "N":
        chocolate_sauce = input("Please choose a valid option to add chocolate sauce - Y for yes, N for no: ")
    if chocolate_sauce.upper() == "Y":
        quantity_chocolate = input("How many pumps do you want? - integers from 1 to 6: ")
        while not quantity_chocolate.isdigit() or int(float(quantity_chocolate))>6 or int(float(quantity_chocolate))<1:
            quantity_chocolate = input("Please input a valid number of pumps - integers from 0 to 6: ")
        extra_charge += 0.5 * (int(float(quantity_chocolate))-2)
        chocolate = "Add "+ str(quantity_chocolate) + " pumps of chocolate"
    else:
        chocolate = "No added chocolate"
    return [chocolate, extra_charge]
```
